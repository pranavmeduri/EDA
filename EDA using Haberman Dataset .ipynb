# Haberman Survival Dataset


Survival of patients who had undergone surgery for breast cancer

The dataset contains cases from a study that was conducted between 1958 and 1970 at the University of Chicago's Billings Hospital on the survival of patients who had undergone surgery for breast cancer.

```python
# IMPORTING HABERMAN'S DATASET

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
haber= pd.read_csv("haberman.csv")
```

# UniVariant Analysis

```python
import numpy as np
survived_more=haber.loc[haber["status"] == 1]
survived_less=haber.loc[haber["status"] == 2]
plt.plot(survived_more["age"],np.zeros_like(survived_more["age"]),'o',label='surv_more')
plt.plot(survived_less["age"],np.zeros_like(survived_less["age"]),'o',label='surv_less')
plt.xlabel('age')
plt.title('Historgram plot')
plt.legend()

plt.show()
```

In the above plot we see that both the surv_more and surv_less are both overlapping together which is very quite hard to understand

# CDF

```python
counts, bin_edges = np.histogram(survived_more['age'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label='surv_more(pdf)');
plt.plot(bin_edges[1:], cdf,label='surv_more(cdf)')


counts, bin_edges = np.histogram(survived_less['age'], bins=20, 
                                 density = True)
pdf = counts/(sum(counts))
plt.plot(bin_edges[1:],pdf,label='surv_less(pdf)');
plt.legend()
plt.title('CDF PLOT')
plt.xlabel('age')
plt.show();
```

Here we see that the percentages for each age of a person of his survmore and surless with the help of the graph

```python
counts, bin_edges = np.histogram(survived_more['age'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label='surv_more(pdf)')
plt.plot(bin_edges[1:], cdf,label='surv_more(cdf)')


# virginica
counts, bin_edges = np.histogram(survived_less['age'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label='surv_less(pdf)')
plt.plot(bin_edges[1:], cdf,label='surv_less(cdf)')
plt.legend()
plt.title('CDF PLOT')
plt.xlabel('age')
```

# Mean, median,Variance, Std

```python
print("Means:")
print(np.mean(survived_more["age"]))
print(np.mean(survived_less["age"]))
print("\nStd-dev:");
print(np.std(survived_more["age"]))
print(np.std(survived_less["age"]))
print("\nMedians:")
print(np.median(survived_more["age"]))
print(np.median(survived_less["age"]))

```
